# Predicting Hypertension: Risikofaktorenanalyse mit NHANES-Daten

#### 1. Hintergrund und Ziel des Projekts:

Hypertonie, also Bluthochdruck, ist ein wichtiger Risikofaktor für Herz-Kreislauf-Erkrankungen. Da Bluthochdruck häufig lange unbemerkt bleibt, ist es aus Public-Health- und Data-Science-Sicht interessant zu untersuchen, welche individuellen Merkmale mit Hypertonie zusammenhängen und ob ein Machine-Learning-Modell den Hypertonie-Status einer Person anhand ausgewählter Risikofaktoren vorhersagen kann.

Die zentrale Fragestellung dieses Projektteils lautet:

**Kann ein Machine-Learning-Modell auf Basis von NHANES-Individualdaten vorhersagen, ob eine Person Hypertonie hat, und welche Risikofaktoren tragen am stärksten zur Vorhersage bei?**

Dabei soll nicht medizinisch diagnostiziert werden, ob eine Person tatsächlich Bluthochdruck hat. Das Ziel ist eine datenbasierte Klassifikation und explorative Analyse auf Basis öffentlich verfügbarer Gesundheitsdaten.

#### 2. Datenquelle

Für dieses Projekt werden Daten aus der National Health and Nutrition Examination Survey (NHANES) verwendet. NHANES ist eine US-amerikanische Gesundheitsstudie, die Interviewdaten, Untersuchungsdaten und Messwerte von Teilnehmenden enthält.

Ausgewählt wurde der Zyklus:

NHANES August 2021–August 2023

Dieser Zeitraum wurde gewählt, weil es sich um einen aktuellen, zusammenhängenden NHANES-Zyklus handelt und die relevanten Datendateien für Blutdruck, Demografie und Körpermaße öffentlich verfügbar sind.

Für den ersten Projektteil werden folgende NHANES-Dateien verwendet:


| **Datei**                 | **Inhalt**                                                                                  | **Rolle im Projekt**                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| `BPQ_L`                   | Blood Pressure & Cholesterol Questionnaire  | Zielvariable Hypertonie.                |
| `DEMO_L`           | Demographic Variables and Sample Weights      | Features: Alter und Geschlecht        |
| `BMX_L`  | Body Measures            | Features: BMI und Taillenumfang

Die Datensätze werden über die gemeinsame Personen-ID SEQN zusammengeführt.


#### 3. Zielvariable / Target

Die Zielvariable beschreibt, ob eine Person Hypertonie hat oder nicht.

Als Target wird zunächst folgende Variable verwendet:

|Variable                 | Bedeutung                                                                                  | Nutzung im Projekt                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| `BPQ020`                   | Ever told you had high blood pressure	|Target für Hypertonie ja/nein

Die Variable BPQ020 fragt, ob einer Person jemals von einem Arzt oder einer anderen medizinischen Fachperson gesagt wurde, dass sie Hypertonie bzw. Bluthochdruck hat.

**Wichtig:** Zur Definition des Hypertonie-Status wurden die Variablen BPQ020, BPQ030 und BPQ150 aus dem Datensatz BPQ_L geprüft. BPQ020 erfasst, ob einer Person jemals Bluthochdruck mitgeteilt wurde. `BPQ030` beschreibt eine wiederholte Mitteilung bei zwei oder mehr Arztbesuchen, während `BPQ150` die aktuelle Einnahme von Blutdruckmedikamenten abbildet.Aufgrund der deskriptiven Prüfung der fehlenden Werte wurden BPQ030 und BPQ150 nicht als Zielvariablen verwendet, da beide einen Missing-Anteil von etwa **65 %** aufwiesen. BPQ020 zeigte dagegen keine fehlenden Werte und wurde daher als binäre Zielvariable für das Klassifikationsmodell ausgewählt.

Für das Machine-Learning-Modell wird diese Variable später in eine binäre Zielvariable umgewandelt:

|Originalwert                | Bedeutung                                                                                  |ML-Codierung                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| 1                   | Ja, Hypertonie wurde mitgeteilt	1	| 1
|2         | Nein, keine Hypertonie mitgeteilt | 0
|7, 9, fehlend | Refused / Don't know / Missing| wird als fehlend behandelt

Bei der Variable `current_smoking_status` lagen viele fehlende Werte vor. Diese wurden nicht direkt als zufällig fehlende Werte interpretiert, sondern im Zusammenhang mit der Fragebogenlogik betrachtet. Die Variable `current_smoking_status` beschreibt den aktuellen Raucherstatus und wird nur Personen gestellt, die zuvor bei `cigarettes_ever_in_life` angegeben haben, bereits mindestens 100 Zigaretten im Leben geraucht zu haben.

Um keinen großen Anteil der Beobachtungen zu verlieren, wurden die fehlenden Werte daher nicht pauschal gelöscht. Stattdessen wurde zusätzlich die Vorfrage `cigarettes_ever_in_life` berücksichtigt. Auf dieser Grundlage wurde eine binäre Variable für den aktuellen Raucherstatus gebildet:

`1 = Ja, aktuell rauchend`
`0 = Nein, aktuell nicht rauchend`

### Behandlung fehlender Werte beim Alkoholkonsum

#### Rekodierung des Alkoholkonsums

Die Variable `alcohol_use` wurde im Rahmen des Feature Engineerings aus den NHANES-Variablen `ever_alcohol_use` und `alcohol_frequency` abgeleitet. Dabei wurde die Fragebogenlogik berücksichtigt. Personen, die angaben, nie Alkohol konsumiert zu haben, sowie Personen ohne Alkoholkonsum im letzten Jahr wurden der Kategorie `no_alcohol` zugeordnet. Die übrigen Angaben zur Konsumhäufigkeit in den letzten zwölf Monaten wurden zu jährlichem, monatlichem, wöchentlichem und täglichem Alkoholkonsum zusammengefasst.

Die Variable `alcohol_use` wurde als ordinale Variable kodiert, wobei höhere Werte eine häufigere Alkoholkonsumfrequenz abbilden:

- 0 = no_alcohol
- 1 = yearly
- 2 = monthly
- 3 = weekly
- 4 = daily

Fehlende Werte, die nach Berücksichtigung der Fragebogenlogik nicht eindeutig zugeordnet werden konnten, wurden mittels Median-Imputation ersetzt. Der Median wurde gewählt, da `alcohol_use` ordinal skaliert ist und dadurch eine robuste zentrale Kategorie zur Ersetzung fehlender Werte verwendet werden kann.

Zusätzlich wurde mit `alcohol_use_label` eine beschreibende Label-Variable erstellt, um die numerischen Kategorien leichter interpretieren zu können.

Für die finale Modellierung wurde zusätzlich die binäre Variable `regular_alcohol_12m` abgeleitet. Dabei wurden wöchentlicher und täglicher Alkoholkonsum in den letzten zwölf Monaten als regelmäßiger Alkoholkonsum kodiert. Kein Alkoholkonsum, jährlicher oder monatlicher Alkoholkonsum wurden als kein regelmäßiger Alkoholkonsum zusammengefasst.


```python
# Sicherstellen, dass die Variablen numerisch sind
df["ever_alcohol_use"] = pd.to_numeric(df["ever_alcohol_use"], errors="coerce")
df["alcohol_frequency"] = pd.to_numeric(df["alcohol_frequency"], errors="coerce")

# Sondercodes als Missing behandeln
df["ever_alcohol_use"] = df["ever_alcohol_use"].replace({
    7: np.nan,
    9: np.nan
})

df["alcohol_frequency"] = df["alcohol_frequency"].replace({
    77: np.nan,
    99: np.nan
})

# alcohol_frequency:
# 0 = Never in last year
# 1 = Every day
# 2 = Nearly every day
# 3 = 3 to 4 times a week
# 4 = 2 times a week
# 5 = Once a week
# 6 = 2 to 3 times a month
# 7 = Once a month
# 8 = 7 to 11 times in last year
# 9 = 3 to 6 times in last year
# 10 = 1 to 2 times in last year
# 77 = Refused
# 99 = Don't know

def create_alcohol_use(row):
    ever_alcohol_use = row["ever_alcohol_use"]
    alcohol_frequency = row["alcohol_frequency"]

    # Person hat nie Alkohol getrunken
    if ever_alcohol_use == 2:
        return 0

    # Person hat im letzten Jahr keinen Alkohol getrunken
    if alcohol_frequency == 0:
        return 0

    # Jährlicher Konsum
    if alcohol_frequency in [8, 9, 10]:
        return 1

    # Monatlicher Konsum
    if alcohol_frequency in [6, 7]:
        return 2

    # Wöchentlicher Konsum
    if alcohol_frequency in [3, 4, 5]:
        return 3

    # Täglicher Konsum
    if alcohol_frequency in [1, 2]:
        return 4

    # echte, nicht erklärbare Missing-Werte
    return np.nan


# Numerische Variable für die Modellierung
df["alcohol_use"] = df.apply(create_alcohol_use, axis=1)

# Lesbare Kategorie für Kontrolle, EDA und App
alcohol_label_map = {
    0: "no_alcohol",
    1: "yearly",
    2: "monthly",
    3: "weekly",
    4: "daily"
}

df["alcohol_use_label"] = df["alcohol_use"].map(alcohol_label_map)


# regular_alcohol_12m erstellen
# 0 = kein regelmäßiger Alkoholkonsum
# 1 = regelmäßiger Alkoholkonsum
#
# weekly und daily werden als regelmäßiger Alkoholkonsum definiert.

df["regular_alcohol_12m"] = np.where(
    df["alcohol_use"].isin([3, 4]), 1, 0
)

```

```python
# Median berechnen
alcohol_median = df["alcohol_use"].median()
df["alcohol_use"] = df["alcohol_use"].fillna(alcohol_median)
df["alcohol_use"] = df["alcohol_use"].astype(int)

# alcohol_use_label nach der Imputation neu erstellen
df["alcohol_use_label"] = df["alcohol_use"].map(alcohol_label_map)
```

```python
print("Median für alcohol_use:", alcohol_median)

print("\nVerteilung alcohol_use:")
print(df["alcohol_use"].value_counts(dropna=False).sort_index())

print("\nVerteilung alcohol_use_label:")
print(df["alcohol_use_label"].value_counts(dropna=False))

print("\nVerteilung regular_alcohol_12m:")
print(df["regular_alcohol_12m"].value_counts(dropna=False))
```

### Behandlung fehlender Werte und Bildung der Variablen zur körperlichen Aktivität und Sitzzeit

Die Angaben zu moderater und intensiver Freizeitaktivität wurden zu einer gemeinsamen Aktivitätsvariable zusammengefasst. Ziel war es, die körperliche Aktivität vereinfacht und gut interpretierbar für die weitere Analyse und Modellierung abzubilden.

Werte größer als 0 wurden als vorhandene Aktivität interpretiert. Sehr kleine numerische Werte nahe 0 wurden als 0 behandelt. Personen ohne moderate ode r intensive Aktivität wurden als nicht aktiv klassifiziert. Personen mit moderater Aktivität, aber ohne intensive Aktivität, wurden als moderat aktiv eingestuft. Sobald intensive Aktivität angegeben wurde, wurde die Person der Kategorie intensiv aktiv zugeordnet.

Die neue Variable `activity_level_label`wurde wie folgt kodiert:

- 0 = nicht aktiv 
- 1 = moderat aktiv
- 2 = intensiv aktiv

Zusätzlich wurde aus der Variable zur täglichen Sitzzeit eine besser interpretierbare metrische Variable gebildet. Da die Sitzzeit in Minuten pro Tag angegeben ist, wurde sie durch 60 geteilt und dadurch in Stunden pro Tag umgerechnet.

Die neu gebildeten Variablen lauten:

`sitting_hours_per_day = Sitzzeit in Stunden pro Tag`

```python

moderate_col = "Moderate-intensity_activities_frequency"
vigorous_col = "high_intensity_activities_frequency"

# sehr kleine Werte praktisch als 0 behandeln
df.loc[df[moderate_col] < 0.000001, moderate_col] = 0
df.loc[df[vigorous_col] < 0.000001, vigorous_col] = 0

# neue Variable erstellen
df["activity_level"] = 0

# moderat aktiv: moderate Aktivität vorhanden, aber keine intensive
df.loc[
    (df[moderate_col] > 0) & (df[vigorous_col] == 0),
    "activity_level"
] = 1

# intensiv aktiv: intensive Aktivität vorhanden
df.loc[
    df[vigorous_col] > 0,
    "activity_level"
] = 2
```

```python
#Für EDA Bereich 
df["activity_level_label"] = df["activity_level"].map({
    0: "nicht aktiv",
    1: "moderat aktiv",
    2: "intensiv aktiv"
})

df["activity_level_label"].value_counts(dropna=False)
```

#### Rekodirung der Variable Sitzzeit

```python
# Sonderwerte als Missing setzen
df["sedentary_time_per_day"] = df["sedentary_time_per_day"].replace({
    7777: np.nan,
    9999: np.nan,
    77: np.nan,
    99: np.nan
})

# Sitzzeit von Minuten in Stunden umrechnen
df["sitting_hours_per_day"] = df["sedentary_time_per_day"] / 60
```

Bei der Prüfung der Sitzzeit in Stunden pro Tag zeigten sich einzelne sehr niedrige Werte unter einer Stunde. Da eine tägliche Sitzzeit von wenigen Minuten im Alltag als unplausibel einzuschätzen ist, wurden Werte unter einer Stunde als fehlende Werte kodiert. Anschließend wurden die fehlenden Werte mithilfe einer Median-Imputation ersetzt.

```python
# sehr kleine / unplausible Sitzzeiten unter 1 Stunde als Missing setzen
df.loc[df["sitting_hours_per_day"] < 1, "sitting_hours_per_day"] = np.nan

# fehlende Werte mit Median ersetzen
df["sitting_hours_per_day"] = df["sitting_hours_per_day"].fillna(
    df["sitting_hours_per_day"].median()
)
```

### Behandlung fehlender Werte bei anthropometrischen Variablen

Da der BMI aus Körpergewicht und Körpergröße berechnet werden kann, wurden fehlende BMI-Werte zunächst anhand der vorhandenen Angaben zu Gewicht und Größe ergänzt. Die Körpergröße lag in Zentimetern vor und wurde für die Berechnung in Meter umgerechnet.

Falls nach dieser Ableitung weiterhin fehlende Werte in den anthropometrischen Variablen vorhanden waren, wurden diese mithilfe einer Median-Imputation ersetzt. Der Median wurde verwendet, da er im Vergleich zum Mittelwert robuster gegenüber Ausreißern ist.

Für die spätere Modellierung werden insbesondere BMI und Taillenumfang als zentrale anthropometrische Risikofaktoren berücksichtigt, da beide Merkmale in einem engen Zusammenhang mit Übergewicht, Adipositas und kardiometabolischem Risiko stehen.

```python
df[["weight(kg)", "height(cm)", "bmi", "waist_circumference(cm)"]].isna().sum()
```

```python
# fehlende werte bei bmi erstmal durch die Rechnung Gewicht / ((Größe_in_cm)/100) ** 2) beahndeln

df['bmi'] = df['bmi'].fillna(
    df['weight(kg)'] / ((df['height(cm)']//100) ** 2)
)
```

```python
# Median-Imputation 

df['waist_circumference(cm)'] = df['waist_circumference(cm)'].fillna(df['waist_circumference(cm)'].median())
df['weight(kg)'] = df['weight(kg)'].fillna(df['weight(kg)'].median())
df['height(cm)'] = df['height(cm)'].fillna(df['height(cm)'].median())
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```

```python
df[["weight(kg)", "height(cm)", "bmi", "waist_circumference(cm)"]].isna().sum()
```

#### Rekodierung von BMI:

Die Standard-Grenzwerte für Erwachsene, von der WHO seit 1995 verwendet und heute von praktisch jeder Gesundheitsbehörde übernommen, lauten: Untergewicht unter 18,5, Normalgewicht 18,5 bis 24,9, Übergewicht 25 bis 29,9 und Adipositas ab 30.

```python
#BMI kategorisieren

df["bmi_category"] = pd.cut(
    df["bmi"],
    bins=[0, 18.5, 24.9, 29.9, 100],
    labels=["underweight", "normal_weight", "overweight", "obesity"],
    right=True
)
```

NHANES-Codierung bei BPQ020 ist wie folgt:
1 = Yes

Für die Zielvariable wurde eine binäre Variable erstellt: 1 = Hypertonie, 0 = keine Hypertonie. Nicht interpretierbare Antworten wie "Refused" oder "Don't know" wurden zunächst nicht zugeordnet und erscheinen dadurch als fehlende Werte.
