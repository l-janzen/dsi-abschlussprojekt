# Predicting Hypertension: Risikofaktorenanalyse mit NHANES-Daten

#### 1. Hintergrund und Ziel des Projekts:

Hypertonie, also Bluthochdruck, ist ein wichtiger Risikofaktor für Herz-Kreislauf-Erkrankungen. Da Bluthochdruck häufig lange unbemerkt bleibt, ist es aus Public-Health- und Data-Science-Sicht interessant zu untersuchen, welche individuellen Merkmale mit Hypertonie zusammenhängen und ob ein Machine-Learning-Modell den Hypertonie-Status einer Person anhand ausgewählter Risikofaktoren vorhersagen kann.

Die zentrale Fragestellung dieses Projektteils lautet:

**Kann ein Machine-Learning-Modell auf Basis von NHANES-Individualdaten vorhersagen, ob eine Person Hypertonie hat, und welche Risikofaktoren tragen am stärksten zur Vorhersage bei?**

Dabei soll nicht medizinisch diagnostiziert werden, ob eine Person tatsächlich Bluthochdruck hat. Das Ziel ist eine datenbasierte Klassifikation und explorative Analyse auf Basis öffentlich verfügbarer Gesundheitsdaten.

#### 2. Datenquelle

Für dieses Projekt werden Daten aus der National Health and Nutrition Examination Survey (NHANES) verwendet. NHANES ist eine US-amerikanische Gesundheitsstudie, die Interviewdaten, Untersuchungsdaten und Messwerte von Teilnehmenden enthält.

Ausgewählt wurde der Zyklus:

NHANES August 2021–August 2023

Dieser Zeitraum wurde gewählt, weil es sich um einen aktuellen, zusammenhängenden NHANES-Zyklus handelt und die relevanten Datendateien für Blutdruck, Demografie und Körpermaße öffentlich verfügbar sind.

Für den ersten Projektteil werden folgende NHANES-Dateien verwendet:


| **Datei**                 | **Inhalt**                                                                                  | **Rolle im Projekt**                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| `BPQ_L`                   | Blood Pressure & Cholesterol Questionnaire  | Zielvariable Hypertonie.                |
| `DEMO_L`           | Demographic Variables and Sample Weights      | Features: Alter und Geschlecht        |
| `BMX_L`  | Body Measures            | Features: BMI und Taillenumfang

Die Datensätze werden über die gemeinsame Personen-ID SEQN zusammengeführt.


#### 3. Zielvariable / Target

Die Zielvariable beschreibt, ob eine Person Hypertonie hat oder nicht.

Als Target wird zunächst folgende Variable verwendet:

|Variable                 | Bedeutung                                                                                  | Nutzung im Projekt                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| `BPQ020`                   | Ever told you had high blood pressure	|Target für Hypertonie ja/nein

Die Variable BPQ020 fragt, ob einer Person jemals von einem Arzt oder einer anderen medizinischen Fachperson gesagt wurde, dass sie Hypertonie bzw. Bluthochdruck hat.

**Wichtig:** Zur Definition des Hypertonie-Status wurden die Variablen BPQ020, BPQ030 und BPQ150 aus dem Datensatz BPQ_L geprüft. BPQ020 erfasst, ob einer Person jemals Bluthochdruck mitgeteilt wurde. `BPQ030` beschreibt eine wiederholte Mitteilung bei zwei oder mehr Arztbesuchen, während `BPQ150` die aktuelle Einnahme von Blutdruckmedikamenten abbildet.Aufgrund der deskriptiven Prüfung der fehlenden Werte wurden BPQ030 und BPQ150 nicht als Zielvariablen verwendet, da beide einen Missing-Anteil von etwa **65 %** aufwiesen. BPQ020 zeigte dagegen keine fehlenden Werte und wurde daher als binäre Zielvariable für das Klassifikationsmodell ausgewählt.

Für das Machine-Learning-Modell wird diese Variable später in eine binäre Zielvariable umgewandelt:

|Originalwert                | Bedeutung                                                                                  |ML-Codierung                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
| 1                   | Ja, Hypertonie wurde mitgeteilt	1	| 1
|2         | Nein, keine Hypertonie mitgeteilt | 0
|7, 9, fehlend | Refused / Don't know / Missing| wird als fehlend behandelt


#### 4. Ausgewählte Features
Die Features wurden ausgewählt, weil sie fachlich nach WHO (Quellenangabe ..... ? )mit Hypertonie in Verbindung stehen, im NHANES-Datensatz verfügbar sind und für ein Klassifikationsmodell gut nutzbar sind.


|Variable               | Bedeutung                                                                                  |Warum ausgewählt?                                                                                 |
|-------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|
|RIDAGEYR                   | Alter in Jahren| Hypertonie tritt mit steigendem Alter häufiger auf
|RIAGENDR         | Geschlecht |Hypertonie-Risiko und Muster können sich zwischen Männern und Frauen unterscheiden
|BMXBMI | Body Mass Index in kg/m² |BMI ist ein wichtiger körperbezogener Risikofaktor
|BMXWAIST |Taillenumfang in cm | Bauchfett bzw. zentraler Körperfettanteil kann mit kardiometabolischem Risiko zusammenhängen

Körpergröß(BMXHT) und Körpergewicht (BMXWT) wurden beibehalten, da sie zur Rekonstruktion fehlender BMI-Werte genutzt werden können. Dadurch kann der Informationsverlust durch fehlende BMI-Angaben reduziert werden.

### 1.2. Auswahl relevanter Variablen

Nach dem Einlesen werden nur die Variablen ausgewählt, die für die Fragestellung relevant sind. Dadurch wird der Datensatz übersichtlicher und die weitere Analyse gezielter. Für die Zielvariable wird aus dem Datensatz BPQ_L die Variable BPQ020 verwendet, da sie angibt, ob einer Person jemals mitgeteilt wurde, dass sie Bluthochdruck hat. Aus DEMO_L werden Alter und Geschlecht ausgewählt, da diese wichtige demografische Einflussfaktoren sind. #### 1.3. Tabellen über SEQN zusammenführen

Nachdem die relevanten Variablen aus den einzelnen NHANES-Dateien ausgewählt wurden, werden die Datensätze über die gemeinsame Personen-ID SEQN zusammengeführt. `SEQN` ist die eindeutige Identifikationsnummer jeder teilnehmenden Person in NHANES. Dadurch können Informationen aus verschiedenen Dateien derselben Person korrekt miteinander verbunden werden.
Das Zusammenführen ist notwendig, weil die relevanten Informationen nicht in einer einzigen Datei gespeichert sind.
Für den Merge wird ein inner join verwendet. Das bedeutet, dass nur Personen im finalen Datensatz bleiben, die in allen ausgewählten Dateien vorhanden sind. Dadurch wird sichergestellt, dass für jede Person alle benötigten Informationen für die erste Analyse- und Modellversion verfügbar sind.

#### 2.5 Behandlung fehlender Werte
Der Umgang mit fehlenden Werten hängt von der betroffenen Variable und vom Anteil fehlender Werte ab. Für die Zielvariable werden fehlende Werte ausgeschlossen, da diese Personen nicht eindeutig einer Hypertonie-Gruppe zugeordnet werden können.

Nach der Rekonstruktion fehlender BMI-Werte wurden Fälle mit weiterhin fehlenden Werten in zentralen Analysevariablen entfernt. Eine Median-Imputation wurde nicht verwendet, da sie künstliche Durchschnittswerte erzeugen kann und individuelle Unterschiede nicht berücksichtigt. Aufgrund des geringen Anteils fehlender Werte wurde eine vollständige Fallanalyse gewählt.

```python
# fehlende Werte von Zielvariable löschen 
df['hypertension'].value_counts(dropna= False)
df = df.dropna(subset= 'hypertension')

# fehlende werte bei bmi erstmal durch die Rechnung Gewicht / ((Größe_in_cm)/100) ** 2) beahndeln dann denn restlichen Na-Werte löschen 
# Fehlende BMI-Werte wurden nicht durch den Median ersetzt, sondern aus Körpergröße und Gewicht rekonstruiert, sofern diese Informationen vorhanden waren. 

df['bmi'] = df['bmi'].fillna(
    df['weight(kg)'] / ((df['height(cm)']//100) ** 2)
)

df = df.dropna(subset= ['bmi', 'weight(kg)', 'height(cm)', 'waist_circumference(cm)'] )

#Eine Median-Imputation wurde nicht gewählt, weil sie allen fehlenden Personen denselben künstlichen Durchschnittswert zuweisen würde.
#Bei medizinischen Variablen wie BMI kann dies die Verteilung und den Zusammenhang mit Hypertonie verzerren.
```
